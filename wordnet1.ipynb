### Using synsets

```python
from nltk.corpus import wordnet as wn

syn = wn.synsets("car")

for s in syn:
    print ("Synset name :  ", s.name(), end = "")
    print ("\nSynset meaning : ", s.definition(), end = "")
    print ("\nSynset example : ", s.examples(), end = "\n\n")
```

### Hypernyms and Hyponyms

```python
syn = wn.synsets('bank')[0]

print ("Synset name :  ", syn.name(), end="")
print ("\nSynset abstract term :  ", syn.hypernyms(), end="")
print ("\nSynset specific term :  ", syn.hypernyms()[0].hyponyms())

syn.root_hypernyms()
print ("\nSynset root hypernerm :  ", syn.root_hypernyms())
```

### Meronyms

```python
car_synset = wn.synset('car.n.01')

print("\n--- Meronyms of 'car.n.01' ---")

# Parts (e.g., wheel, engine)
parts = car_synset.part_meronyms()
print("Part Meronyms:", parts)
```

### Holonyms

```python
# Example Synset for 'wheel' to find its holonyms
wheel_synset = wn.synset('wheel.n.01')
print("\n--- Holonyms of 'wheel.n.01' ---")
part_of_holonyms = wheel_synset.part_holonyms()
print("Part Holonyms:", part_of_holonyms)


# Example Synset for 'tree' to find a member holonym
tree_synset = wn.synset('tree.n.01')
member_of_holonyms = tree_synset.member_holonyms()
print(f"\n--- Holonyms of 'tree.n.01' ---")
print("Member Holonyms:", member_of_holonyms)
```

### Similarity between two words

```python
dog = wn.synsets('dog')[0]
cat = wn.synsets('cat')[0]
sim = dog.path_similarity(cat)

print ("\nSimilarity between dog and cat :  ", sim)
```

```python
one = 'cat'
other = 'car'

entity1 = wn.synsets(one)
entity2 = wn.synsets(other)

for x in entity1:
    for y in entity2:
        sim = x.path_similarity(y)
        print ("\nSimilarity between " + x.name() + " and " + y.name() + " :  ", sim)
```

### POS Tagging

```python
syn = wn.synsets('cook')
print ("Syn tag : ", [x.name() for x in syn])

syn = wn.synsets('doing')[0]
print ("Syn tag : ", syn.pos())

syn = wn.synsets('beautiful')[0]
print ("Syn tag : ", syn.pos())

syn = wn.synsets('quickly')[0]
print ("Syn tag : ", syn.pos())
```
